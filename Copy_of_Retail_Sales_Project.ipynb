```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales.csv")

df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sales_by_product = df.groupby(
    'Product'
)['Sales'].sum()

print(sales_by_product)
```

```python
sales_by_product.plot(
    kind='bar'
)

plt.title("Total Sales by Product")

plt.xlabel("Product")

plt.ylabel("Sales")

plt.show()
```

```python
sns.lineplot(
    x='Month',
    y='Sales',
    hue='Product',
    data=df
)

plt.title("Monthly Sales Trend")

plt.show()
```

```python
plt.scatter(
    df['Advertising'],
    df['Sales']
)

plt.xlabel("Advertising")

plt.ylabel("Sales")

plt.title("Advertising vs Sales")

plt.show()
```

```python
corr = df[['Advertising','Sales']].corr()

corr
```

```python
sns.heatmap(
    df[['Advertising','Sales']].corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```
